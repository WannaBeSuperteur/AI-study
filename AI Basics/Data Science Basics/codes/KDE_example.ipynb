```python
import numpy as np
import pandas as pd
import plotly

n_data = 50
data = np.random.normal(0, 5, n_data)
print(data)
```

```python
def gaussian(x):
    return (1.0 / np.sqrt(2.0 * np.pi) / np.exp(x**2.0))

def box(x):
    if abs(x) <= 0.5:
        return 1.0
    else:
        return 0.0

def tri(x):
    if abs(x) <= 1.0:
        return 1.0 - abs(x)
    else:
        return 0.0
```

```python
import plotly.graph_objects as go

x = np.linspace(-1.5, 1.5, 300)
y_gaussian = np.vectorize(gaussian)(x)
y_box = np.vectorize(box)(x)
y_tri = np.vectorize(tri)(x)

fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y_tri, mode='lines', name='y = sin(x)'))

# 그래프 레이아웃 설정
fig.update_layout(
    xaxis_title="x",
    yaxis_title="y"
)
fig.update_layout(width=400, height=300)

# 그래프 표시
fig.show()
```

```python
import plotly.express as px

original_gaussian = [0.0 for i in range(2001)]
values_gaussian = [0.0 for i in range(2001)]
values_box = [0.0 for i in range(2001)]
values_tri = [0.0 for i in range(2001)]

for d in data:
    for i in range(len(values_gaussian)):
        original_gaussian[i] += gaussian((i - 1000.0) / 500.0) / 5 # / n_data
        values_gaussian[i] += gaussian(((i - 1000.0) / 100.0) - d) # / n_data
        values_box[i] += box(((i - 1000.0) / 100.0) - d) # / n_data
        values_tri[i] += tri(((i - 1000.0) / 100.0) - d) # / n_data

x = np.linspace(-10.0, 10.0, 2001)
fig = px.histogram(pd.DataFrame({'histogram': data}), nbins=50, opacity=0.667, labels=['histogram'])
fig.add_trace(go.Scatter(x=x, y=original_gaussian, mode='lines', name='Original Gaussian'))
fig.add_trace(go.Scatter(x=x, y=values_gaussian, mode='lines', name='Gaussian Kernel Prediction'))
# fig.add_trace(go.Scatter(x=x, y=values_box, mode='lines', name='Box'))
# fig.add_trace(go.Scatter(x=x, y=values_tri, mode='lines', name='Tri'))
fig.update_layout(bargap=0.2)

# 그래프 레이아웃 설정
fig.update_layout(
    xaxis_title="x",
    yaxis_title="y"
)
for d in data:
    fig.add_scatter(x=[d], y=[0.1],
                    marker=dict(
                        color='black',
                        size=4
                    ))
fig.update_layout(width=800, height=500)

# 그래프 표시
fig.show()
```
